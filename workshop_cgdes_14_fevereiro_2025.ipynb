### Média

A média é definida como sendo a soma de todos os elementos ($x_1 + x_2 + ... + x_n$) dividida pelo número de elementos da amostra ($n$).

\begin{align}
Média = \overline{X} = \frac{x_1 + x_2 + ... + x_n}{n} = \frac{\sum_{i=1}^n x_i}{n}
\end{align}

```python
# Gerar 20 números randômicos entre 1 e 100
randomList = random.sample(range(1, 60), 6)
print(randomList)
```

```python
# Determinar a média
sum = 0
for i in randomList:
    sum += i
print(f"Soma                = {sum}")
print(f"Quantidade de dados = {len(randomList)}")
media = sum/len(randomList)
print(f"Média               = {media}")
```

### Mediana

Representa o valor central de um conjunto de dados ordenados (crescente ou decrescente). Quando o número de itens é par, calcula-se a média dos dois elementos centrais.

```python
# Determinar a mediana
sortedList = sorted(randomList)
print(sortedList)
if len(sortedList)%2 == 0: # qtd par de itens
    # soma os 2 itens do meio e divide por 2
    mediana = (sortedList[int(len(sortedList)/2)]+sortedList[int(len(sortedList)/2-1)])/2
else:
    # divide o item do meio por 2
    mediana = sortedList[int(len(sortedList)/2)]
print(f"Mediana   = {mediana}")
```

### Desvio padrão

Medida de dispersão em torno da média. Um desvio padrão baixo indica que os valores de um conjunto estão próximos da média. Em resumo, o desvio padrão é uma medida que indica a dispersão dos dados de uma amostra em relação à média.

O desvio padrão é dependente da variância em relação à média. Então, o primeiro passo é calcular essa variância subtraindo cada uma das medidas individuais pela média e em seguida elevar os resultados obtidos ao quadrado. O próximo passo é calcular a raiz quadrada da soma de todos resultados e dividos pelo tamanho da amostra.

\begin{align}
    Desvio~padrão = \sigma = \sqrt{\frac{1}{n} \sum_{i=1}^n (x_i-\overline{X})^2 }
\end{align}
    
* *Para o desvio-padrão populacional divide-se pelo número da população $n$ e para o desvio-padrão amostral dividimos pelo número da amostra menos um $n-1$.*

```python
# Determinar o desvio padrão
sumErrors = 0
for i in range(len(sortedList)):
    sumErrors += (sortedList[i]-media)**2
stdDev = (sumErrors/(len(sortedList)))**(0.5)
print(f"Desvio padrão  = {stdDev}")
```

### Erro padrão da média

O desvio padrão é calculado a partir de uma amostra. Então, é razoável supor que se a amostra fosse diferente, o valor do desvio padrão também podria ser diferente. Mas quão diferente seria? Para responder a esta qiestão, calcula-se o erro padrão da média.

O erro padrão mede a variação de uma média amostral em relação à média da população. A estimativa do erro padrão pode ser calculada dividindo o desvio padrão pela raiz quadrada do tamanho amostral.

\begin{align}
       Erro~padrão =  \sigma_\overline{X} = \frac{\sigma}{\sqrt{n}}
    \end{align}

```python
stdError = stdDev/(len(sortedList)**0.5)
print(f"Erro padrão da média = {stdError}")
```

## Aprendizado supervisionado

Em aprendizado de máquinas, o aprendizado supervisionado é aquele em que um conjunto de dados conhecido e "etiquetado" é usado para fazer predições.

Os algoritmos de aprendizado supervisionado constroem modelos capazes de fazer predições em um novo conjunto de dados a partir da análise de duas variáveis do dataset: entrada (X) e saída (Y).

## Regressão

A regressão é utilizada para verificar se há relação entre uma variável resposta (variável dependente) e uma ou mais variáveis independentes.


## Regressão linear

Aproximação linear para modelar a relação entre uma variável dependente e uma ou mais variáveis independentes.

Para 1 variável independente, temos regressão linear simples. Para duas ou mais variáveis independentes, temos regressão linear múltipla.


Seja $Y = mX + C$, onde $m$ é o coeficiente angular da reta e $C$ é o coeficiente linear. O objetivo é determinar os valores de C que  produzem o menor erro utilizando o método dos mínimos quadrados.

Para minimizar o erro, primeiro precisamos calculá-lo com uma função de perda (*loss function*). A função *quadratic loss function* é dada por:
\begin{align}
loss~function  = L(x) = \sum_{i=1}^n(Y_i-\overline{Y}_i)^2
\end{align}

Onde $Y_i$ é o valor real e $\overline{Y}$ o valor previsto. O $^2$ é para evitar valores negativos no erro.

Uma vez definida a função de perda, é preciso minimizá-la através de derivadas parciais de $L$ e igualando a zero. Obtendo assim as seguintes expressões:

\begin{align}
m = \frac{\sum_{i=1}^n(X_i-\overline{X})(I_i-\overline{Y})}{\sum_{i=1}^n(X_i-\overline{X})^2}
\end{align}
\begin{align}
\\
\\
\end{align}
\begin{align}
C = \overline{Y} - m\overline{X}
\end{align}

onde $\overline{X}$ é o valor médio de $X$ e $\overline{Y}$ é o valor médio de $Y$

```python
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# Carregando os dados nas variáveis diabetes_X e diabetes_y
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

```python
diabetes_X.shape, diabetes_y.shape
```

```python
# Selecionando apenas uma característica do dataset: índice de massa corporal (body mass index) ou bmi
diabetes_X = diabetes_X[:, np.newaxis, 2]
```

```python
diabetes_X.shape # somente bmi
```

```python
diabetes_X[1:2]
```

```python
diabetes_y[1:2]
```

Separando os dados em dois conjuntos training/testing (80%, 20%)

```python
diabetes_X_train = diabetes_X[:-84] # 0 até n-84
diabetes_X_test = diabetes_X[-84:] # n-84 até n

diabetes_y_train = diabetes_y[:-84] # 0 até n-84
diabetes_y_test = diabetes_y[-84:] # n-84 até n
```

```python
print(f"Total de amostras de índice de massa corporal: {diabetes_X.shape[0]}")
print(f"Amostras de índice de massa corporal para treino: {diabetes_X_train.shape[0]}")
print(f"Amostras de índice de massa corporal para teste: {diabetes_X_test.shape[0]}")
print(f"Indicadores de diabetes para treino: {diabetes_y_train.shape[0]}")
print(f"Indicadores de diabetes para teste: {diabetes_y_test.shape[0]}")
```

Criando um objeto de regressão linear (modelo)

```python
modelo_regressao_linear_diabetes = linear_model.LinearRegression()
```

Treinando o modelo com os dados de treino (diabetes_X_train, diabetes_y_train)

O modelo é treinado a partir da amostra do indicador de diabetes (y) em função do índice de massa corporal (X).

Dessa forma, o modelo traça uma reta (regressão linear) a partir do pontos do plano cartesiano. Essa reta representa a distribuição dos dados.

```python
modelo_regressao_linear_diabetes.fit(diabetes_X_train, diabetes_y_train)
```

Predizendo a incidência de diabetes no conjunto de teste (X) a partir do modelo treinado.

Com base no aprendizado com *X_train*, o objetivo é contruir um *y_test* a partir de um X_test.

Note que o seu modelo já conhece os dados de *X_train*, mas nunca "viu" os dados de *X_test*.

Assim, o modelo tenta inferir um *y_test* para um *X_test* a partir do que aprendeu com *X_train* e *y_train*.

```python
diabetes_y_pred = modelo_regressao_linear_diabetes.predict(diabetes_X_test.reshape(-1, 1))
```

Plot do resultado da regressão linear

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(10,7.5))
plt.scatter(diabetes_X_train, diabetes_y_train, color='red', alpha=.5, label='Conjunto de treino')
plt.scatter(diabetes_X_test, diabetes_y_test, color="green", alpha=.5, label='Conjunto de teste')
plt.plot(diabetes_X_test, diabetes_y_pred, color='blue', label='Modelo')
plt.xlabel('Índice de massa corporal')
plt.ylabel('Incidência de diabetes')
# plt.xticks(())
# plt.yticks(())
plt.legend(loc='best')
plt.show()
```

Avaliando a predição com o coeficiente de determinação: 1 é uma predição perfeita

```python
from sklearn.metrics import mean_squared_error, r2_score
print('Coeficiente de determinação: %.2f' % r2_score(diabetes_y_test, diabetes_y_pred))
```

Essa situação é conhecida como *underfitting*. Da mesma forma, um modelo que se ajuste perfeitamente a um conjunto de dados, não poderá ser usado de forma genérica, ocorrendo o *overfitting*. Neste caso podemos usar a regressão polinomial para determinar uma curva que explique melhor que uma reta o comportamento dos dados.

## Classificação

Em aprendizado de máquina, a classificação é a tarefa de agrupamento dos dados em diferentes classes, grupos ou partes.

Por exemplo, se tomarmos os dados de vitórias, empates ou derrotas de um jogador de bilhar ao longo do tempo, podemos classificá-lo em um dado período como "em forma" ou "fora de forma" de acordo com seu desempenho. Uma vez que o modelo aprende a classificar com este jogador, é possível aplicar o mesmo modelo a outros jogadores de bilhar.

### Classificação binária

É o processo de classificação no qual elementos de um conjunto de dados são classificados em 2 classes. Um dos exemplos mais recorrentes é a classificação de e-mails como "*spam*" ou "não *spam*" com base em palavras-chave. A partir de um conjunto de textos de e-mail previamente etiquetados como *spam* ou não, constrói-se um modelo capaz de reconhecer características do texto e separar o texto no conjunto *spam* ou não *spam*. Uma vez construído o modelo, ele pode ser aplicado a qualquer outro texto para classificá-lo como *spam* ou não.


### Classificação multiclasses

É o processo de classificação no qual elementos de um conjunto de dados são classificados em $n$ classes. Um exemplo recorrente é a classificação de espécies de plantas com base em imagens.


### Avaliação da classificação

Algumas métricas são utilizadas para avaliar o quão bom um modelo é capaz de classificar.

#### Acurácia

* True positive (TP): quando um elemento pertence a uma classe e o modelo diz que ele pertence.
* False positive (FP): quando o elemento não pertence a uma classe e o modelo diz que ele pertence;
* True negative (TN): quando um elemento não pertence a uma classe e o modelo diz que ele não pertence;
* False negative (FN): quando um elemento pertence a uma classe e o modelo diz que ele não pertence;
    
A acuária é obtida como se segue:
    
$$ acurária = \frac{TP + TN}{TP + FP + TN + FN} $$

A acurácia pode ser representada por uma matriz de confusão.

#### Precisão

A precisão é a métrica que indica qual a proporção de identificações positivas foi correta. Ou seja, entre todas as identificações positivas, em que proporção o modelo realmente acertou. A precisão é dada pela fórmula:

$$ Precisão = \frac{TP}{TP + FP} $$


#### Recall

O *recall* indica a proporção de verdadeiros positivos corretamente identificados. A fórmula para cacular o *recall* é dada por:

$$ Recall = \frac{TP}{TP + FN} $$

## Máquina de vetores de suporte (Support Vector Machine - SVM)

É um algoritmo supervisionado de aprendizado de máquina cujo objetivo é encontrar um hiperplano num espaço $n$-dimenisional, onde $n$ é o número de características que distingue os pontos. O hiperplano escolhido deve ser aquele com margem máxima, ou seja, máxima distância entre os pontos das classes. Pode ser usado para problemas de regressão, mas principalmente para problemas de classificação. A Figura abaixo, da documentação do Sci-Kit-Learn mostra um exemplo de hiperplano.

### Algumas considerações sobre parâmetros SVM

O kernel pode ser linear, polinomial, gaussiano, RBF e sigmoid.

O parâmetro de regularização (C) determina o tamanho da margem.

* Valores altos $\rightarrow$ margem pequena.
* Valores baixos $\rightarrow$ margem larga.

O parâmetro gamma indica quais pontos devem ser considerados em relação à margem.

* Gamma alto $\rightarrow$ apenas pontos próximos à margem são considerados para classificação;
* Gamma baixo $\rightarrow$ pontos distantes da margem são considerados para classificação.

Em termos práticos, o parâmetro C, comum a todos os kernels SVM, compensa a classificação incorreta dos exemplos de treinamento pela simplicidade da superfície de decisão. Um valor baixo de C torna a superfície de decisão suave, enquanto um valor alto de C visa classificar todos os exemplos de treinamento corretamente.

O parâmetro gamma define quanta influência um único exemplo de treinamento tem. Quanto maior for o gamma, mais próximos os outros exemplos devem estar para serem afetados.

**A escolha adequada de C e gamma é crítica para o desempenho do SVM.**

```python
# Kernel linear, parâmetro de regularização = 1000, gamma = 1
svm_linear = svm.SVC(kernel = 'linear', C = 1000, gamma = 1)
svm_linear.fit(X, y)
```

```python
# Plot
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

ax   = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# grid
xx     = np.linspace(xlim[0], xlim[1], 30)
yy     = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy     = np.vstack([XX.ravel(), YY.ravel()]).T
Z      = svm_linear.decision_function(xy).reshape(XX.shape)

# fronteiras
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
# vetores de suporte
ax.scatter(
    svm_linear.support_vectors_[:, 0],
    svm_linear.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors='none',
    edgecolors='k'
)
plt.show()
```

Ums função sigmóide é dada por:

$$ \sigma = s(x) = \frac{1}{1+e^{-x}} $$

Ela modela uma resposta linear em um valor de probabilidade entre $0$ e $1$.

```python
x_sig = np.linspace(-10, 10, 50)
y_sig = 1/(1 + np.exp(-x_sig))
plt.plot(x_sig, y_sig, color = '#2D46B9', marker = "o")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Função Sigmóide")
plt.show()
```

A função $tanh$ reescala a função sigmóide para um inter valor entre $-1$ e $1$ e é dada por:

$$ tanh(x) = 2\sigma(2x)-1 $$

```python
x_sig = np.linspace(-10, 10, 50)
y_sig = np.tanh(x_sig)
plt.plot(x_sig, y_sig, color = '#2D46B9', marker = "o")
plt.title("Função Tanh")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## Árvores de Decisão

Uma árvore é uma coleção de elementos chamados **nós**, sendo um deles uma **raíz**.
A ligação entre os nós implica em uma uma relação de *paternidade* e impõe a eles uma estrutura hierárquica.

O objetivo da árvore de decisão é diminuir a entropia dos dados de forma inversamente proporcional ao ganho de informação. A entropia é uma medida de incerteza ou aleatoriedade.
Ou seja, quanto maior a entropia, maior a desorganização.

Ao reorganizar os dados de uma determinada maneira, uma árvore de decisão busca diminuir a entropia e há um ganho de informação.

A árvore de decisão testa de forma recursiva todas as possibilidades de organização a cada nível para determinar aquela onde há maior ganho de informação e consequentemente, menor entropia.

### Como a árvore de decisão escolhe a *feature* que mais diminui a entropia?

Material retirado do *Crash Course* de Inteligência Artificial para todos do Diogo Cortiz.

Considere o seguinte conjunto de dados:

| Idade 	| Sexo 	| Pressão arterial 	| Colesterol 	| Droga 	|
|-------	|------	|------------------	|------------	|-------	|
| 23    	| F    	| Alta             	| Alto       	| A     	|
| 47    	| M    	| Baixa            	| Alto       	| A     	|
| 47    	| M    	| Baixa            	| Alto       	| B     	|
| 28    	| F    	| Normal           	| Alto       	| B     	|
| 61    	| F    	| Baixa            	| Alto       	| B     	|
| 22    	| F    	| Normal           	| Alto       	| A     	|
| 49    	| F    	| Normal           	| Alto       	| B     	|
| 41    	| M    	| Baixa            	| Alto       	| A     	|
| 60    	| M    	| Normal           	| Alto       	| B     	|
| 43    	| M    	| baixa            	| Normal     	| A     	|


A pergunta de pesquisa seria: como determinar qual droga usar, A ou B para um novo caso a partir dos dados apresentados acima?

O primeiro nó da árvore (raiz) terá todas as possibilidades de drogas do *dataset*.

| A 	| A 	| B 	| B 	| B 	| A 	| B 	| A 	| B 	| A 	|
|---	|---	|---	|---	|---	|---	|---	|---	|---	|---	|

São 5 possibilidades da droga **A** e cinco possibilidades da droga **B**

\begin{align}
Entropia  = - \sum_{i}P_i~log_2~P_i
\end{align}

\begin{align}
Entropia  = -(P(A)*log_2~P(A))~-(P(B)*log_2~P(B))
\end{align}

\begin{align}
P(A)=\frac{5}{10}=0.5\\
P(B)=\frac{5}{10}=0.5\\
\end{align}

\begin{align}
Entropia  = -(0.5*log_2~0.5)~-(0.5*log_2~0.5)
\end{align}

\begin{align}
Entropia  = 1
\end{align}
* A entropia pode ser $>1$ em sistemas multiclasses.

A partir desse ponto, está determinada a entropia inicial. Agora a árvore de decisão testará todas as *features* e verificará qual delas diminui mais a entropia, resultando em maior ganho de informação.

Exemplo: Se a *feature* escolhida for **Sexo**, a raiz será dividida em 2 nós, com os seguintes vaores:

Nó **F**: [A,A,B,B,B]

Nó **M**: [A,A,A,B,B]


\begin{align}
Ganho  = Entropia(pai)~-\sum_{i}peso(filhos)*Entropia(filhos), onde\\
peso(filhos) = \frac{nó~filho}{nó~pai}
\end{align}

Sabendo que a entropia dos filhos é:
\begin{align}
Entropia(F)=0.97\\
Entropia(M)=0.97
\end{align}

Substituindo na fórmula do Ganho:
\begin{align}
Ganho  = 1~-(0.5*0.97)+(0.5*0.97)
\end{align}

\begin{align}
Ganho = 0.03
\end{align}

Após testar todas a *features*, aquela divisão de dados que gerar maior ganho, permanece.

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
%pylab inline
```

##### Lendo e tratando os dados do dataset

```python
url  = "https://raw.githubusercontent.com/giiff/livro_algoritmos_e_estruturas_de_dados/master/csv_files/CarPrice.csv"
carros = pd.read_csv(url)
carros = carros.drop('car_ID', axis = 1)
carros['marca'] = carros['CarName'].str.split(" ", n=1, expand=True)[0]
carros['marca'] = carros['marca'].replace('alfa-romero','alfa-romeo')
carros['marca'] = carros['marca'].replace('maxda','mazda')
carros['marca'] = carros['marca'].replace('Nissan','nissan')
carros['marca'] = carros['marca'].replace('porcshce','porsche')
carros['marca'] = carros['marca'].replace('vokswagen','volkswagen')
carros['marca'] = carros['marca'].replace('vw','volkswagen')
carros['marca'] = carros['marca'].replace('toyouta','toyota')
carros.columns = ['risco', 'modelo', 'combustível', 'aspiração', 'qtd_portas', 'tipo',
       'tração', 'posicao_motor', 'distancia_entre_eixos', 'comprimento', 'largura',
       'altura', 'peso_cheio', 'tipo_motor', 'cilindros', 'tamanho_motor',
       'sistema_combustível', 'diametro_cilindro', 'volume_cilindro', 'taxa_compressao', 'cavalos_potencia',
       'pico_rpm', 'km_na_cidade', 'km_na_estrada', 'preço', 'marca']
carros['km_na_cidade'] = carros.apply(lambda x : x['km_na_cidade']*0.62137, axis = 1)
carros['km_na_estrada'] = carros.apply(lambda x : x['km_na_estrada']*0.62137, axis = 1)
carros['modelo'] = carros.apply(lambda x : " ".join(x['modelo'].split(" ")[1:]), axis = 1)
carros['modelo'] = carros.marca.map(str) + " - " + carros.modelo
carros['modelo'] = carros['modelo'].replace('audi - 100 ls','audi - 100ls')
```

```python
def strNumber_to_intNumber( n ):
    if n.strip() == 'one': return 1
    if n.strip() == 'two': return 2
    if n.strip() == 'three': return 3
    if n.strip() == 'four': return 4
    if n.strip() == 'five': return 5
    if n.strip() == 'six': return 6
    if n.strip() == 'seven': return 7
    if n.strip() == 'eight': return 8
    if n.strip() == 'nine': return 9
    if n.strip() == 'ten': return 10
    return n

carros['qtd_portas'] = carros['qtd_portas'].apply(lambda x : strNumber_to_intNumber(str(x)))
carros['cilindros'] = carros['cilindros'].apply(lambda x : strNumber_to_intNumber(str(x)))
```

```python
carros
```

```python
labels = pd.DataFrame()
labels['modelo'] = carros['modelo']
labels['id'] = pd.factorize(labels.modelo)[0]
```

```python
# Convertendo valores textuais em numéricos
carros.combustível = pd.factorize(carros.combustível)[0]
carros.aspiração = pd.factorize(carros.aspiração)[0]
carros.qtd_portas = pd.factorize(carros.qtd_portas)[0]
carros.tipo = pd.factorize(carros.tipo)[0]
carros.tração = pd.factorize(carros.tração)[0]
carros.posicao_motor = pd.factorize(carros.posicao_motor)[0]
carros.tipo_motor = pd.factorize(carros.tipo_motor)[0]
carros.cilindros = pd.factorize(carros.cilindros)[0]
carros.sistema_combustível = pd.factorize(carros.sistema_combustível)[0]
carros.marca = pd.factorize(carros.marca)[0]
carros.modelo = pd.factorize(carros.modelo)[0]
```

```python
carros = carros.dropna() # removendo eventuais registros com NA para evitar distorções
carros
```

```python
# Função retirada de Python Data Science Handbook, author: Jake VanderPlas
# https://jakevdp.github.io/PythonDataScienceHandbook/
def classificar_e_visualizar(model, X, y, ax=None):
    ax = ax or plt.gca()
    # Plot dos pontos de treinamento
    ax.scatter(X[:, 0], X[:, 1], c=y, s=10, cmap='viridis',
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # Treinar e classificar
    model.fit(X, y)
    xx, yy = np.meshgrid(np.linspace(*xlim, num=200),
                         np.linspace(*ylim, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    # Plot resultados com área de classificação delimitada
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.2,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap='viridis',
                           zorder=1)

    ax.set(xlim=xlim, ylim=ylim)
```

```python
%pylab inline
pylab.figure(figsize=(15,15))
classificar_e_visualizar(RandomForestClassifier(), X, y);
```

```python
%pylab inline
pylab.figure(figsize=(15,15))
from sklearn.gaussian_process import GaussianProcessClassifier
classificar_e_visualizar(GaussianProcessClassifier(), X, y);
```
